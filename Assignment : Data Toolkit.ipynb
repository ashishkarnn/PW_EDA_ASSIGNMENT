```python
# Question: Demonstrate three different methods for creating identical 2D arrays in NumPy.
# Provide the code for each method and the final output after each method.

import numpy as np

# Method 1: Using np.array with a nested list
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Method 1 - Using np.array:")
print(array1)

# Method 2: Using np.reshape on a range
array2 = np.arange(1, 7).reshape(2, 3)
print("\nMethod 2 - Using np.arange and reshape:")
print(array2)

# Method 3: Using np.fromiter and reshape
array3 = np.fromiter(range(1, 7), dtype=int).reshape(2, 3)
print("\nMethod 3 - Using np.fromiter and reshape:")
print(array3)
```

```python
# Question 3: Explain the difference between np.array, np.asarray, and np.asanyarray.

# These functions are used to convert data to NumPy arrays:
# np.array() always creates a copy.
# np.asarray() avoids copy if input is already an array.
# np.asanyarray() behaves like asarray but keeps subclass type (e.g., matrix).

arr = [1, 2, 3]
a1 = np.array(arr)
a2 = np.asarray(arr)
a3 = np.asanyarray(arr)

print(f"np.array: {type(a1)}")
print(f"np.asarray: {type(a2)}")
print(f"np.asanyarray: {type(a3)}")
```

```python
# Question 4: The difference between deep copy and shallow copy.

a = np.array([1, 2, 3])
b_shallow = a
b_deep = a.copy()

a[0] = 99

print("Original Array (a):", a)
print("Shallow Copy (b_shallow):", b_shallow)
print("Deep Copy (b_deep):", b_deep)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
dates = pd.date_range(start='2023-01-01', periods=100)
temperature = np.random.uniform(20, 30, size=100)
humidity = np.random.uniform(40, 60, size=100)
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Create the plot with two y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], label='Temperature', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], label='Humidity', color='red')
ax2.set_ylabel('Humidity', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Set the title and add legends
ax1.set_title('Temperature and Humidity Over Time')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate 1000 samples from a standard normal distribution
data = np.random.normal(size=1000)

# Plot the histogram with density=True for probability density
plt.hist(data, bins=30, density=True, alpha=0.6, label='Histogram')

# Calculate and plot the PDF
x = np.linspace(min(data), max(data), 100)
pdf = stats.norm.pdf(x)
plt.plot(x, pdf, 'r', label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Assign quadrants based on position
quadrant = []
for i in range(100):
    if x[i] > 0 and y[i] > 0:
        quadrant.append('I')
    elif x[i] < 0 and y[i] > 0:
        quadrant.append('II')
    elif x[i] < 0 and y[i] < 0:
        quadrant.append('III')
    elif x[i] > 0 and y[i] < 0:
        quadrant.append('IV')
    else:
        quadrant.append('Origin')

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrant})

# Create the Seaborn scatter plot
sns.scatterplot(data=df, x='x', y='y', hue='quadrant')

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='Y')

# Add the line (grid lines are included by default)
p.line(x, y)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

# Generate categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=5)
colors = ['green' if v < 3 else 'yellow' if v < 6 else 'red' for v in values]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values, colors=colors))

# Create a Bokeh figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values')

# Add vertical bars
p.vbar(x='categories', top='values', width=0.9, color='colors', source=source)

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Show the plot
show(p)
```
