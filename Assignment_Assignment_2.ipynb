```python


# 1. **What is hypothesis testing in statistics?**
#    Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample to infer that a certain condition holds for the entire population.

# 2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
#    - **Null Hypothesis (H₀)**: Assumes no effect or no difference. Example: "The average height of students is 5.5 ft."
#    - **Alternative Hypothesis (H₁)**: Assumes an effect or difference exists. Example: "The average height of students is not 5.5 ft."

# 3. **What is the significance level in hypothesis testing, and why is it important?**
#    The significance level (α) is the probability of rejecting the null hypothesis when it is true. Common levels are 0.05 or 0.01. It determines the threshold for deciding if results are statistically significant.

# 4. **What does a P-value represent in hypothesis testing?**
#    The P-value is the probability of observing data as extreme as the sample, assuming the null hypothesis is true. Lower P-values indicate stronger evidence against H₀.

# 5. **How do you interpret the P-value in hypothesis testing?**
#    - If \( P \leq \alpha \): Reject H₀ (evidence supports H₁).
#    - If \( P > \alpha \): Fail to reject H₀ (not enough evidence for H₁).

# 6. **What are Type 1 and Type 2 errors in hypothesis testing?**
#    - **Type 1 Error**: Rejecting H₀ when it is true. Probability = α.
#    - **Type 2 Error**: Failing to reject H₀ when it is false. Probability = β.

# 7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
#    - **One-tailed Test**: Tests for an effect in one direction. Example: "Mean is greater than 5."
#    - **Two-tailed Test**: Tests for an effect in both directions. Example: "Mean is not equal to 5."

# 8. **What is the Z-test, and when is it used in hypothesis testing?**
#    The Z-test compares sample means to known population means when the population variance is known and the sample size is large (\( n > 30 \)).

# 9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**
#    \( Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}} \), where \( \bar{x} \) is the sample mean, \( \mu \) is the population mean, \( \sigma \) is the population standard deviation, and \( n \) is the sample size. It represents how many standard deviations the sample mean is from the population mean.

# 10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
#     The T-distribution is used when the sample size is small (\( n \leq 30 \)) and the population standard deviation is unknown.

# 11. **What is the difference between a Z-test and a T-test?**
#     - **Z-test**: Requires known population variance, suitable for large samples.
#     - **T-test**: Suitable for small samples or unknown population variance.

# 12. **What is the T-test, and how is it used in hypothesis testing?**
#     The T-test compares means between groups or a sample mean to a population mean. It is used for small samples and unknown population variance.

# 13. **What is the relationship between Z-test and T-test in hypothesis testing?**
#     Both tests evaluate means, but the T-test accounts for uncertainty in variance for small samples, while the Z-test assumes known variance.

# 14. **What is a confidence interval, and how is it used to interpret statistical results?**
#     A confidence interval provides a range of values within which the true parameter is likely to lie with a certain confidence level (e.g., 95%).

# 15. **What is the margin of error, and how does it affect the confidence interval?**
#     The margin of error is the maximum expected difference between the sample statistic and the population parameter. Larger margins widen the confidence interval.

# 16. **How is Bayes' Theorem used in statistics, and what is its significance?**
#     Bayes' Theorem calculates conditional probabilities and updates beliefs based on new evidence. It is significant in decision-making under uncertainty.

# 17. **What is the Chi-square distribution, and when is it used?**
#     The Chi-square distribution is used for categorical data to test independence or goodness of fit.

# 18. **What is the Chi-square goodness-of-fit test, and how is it applied?**
#     It determines if observed frequencies match expected frequencies. Application involves calculating \( \chi^2 = \sum \frac{(O - E)^2}{E} \) and comparing it to a critical value.

# 19. **What is the F-distribution, and when is it used in hypothesis testing?**
#     The F-distribution compares variances of two samples and is used in ANOVA and F-tests.

# 20. **What is an ANOVA test, and what are its assumptions?**
#     ANOVA (Analysis of Variance) tests if means of multiple groups are equal. Assumptions:
#     - Normality of data,
#     - Homogeneity of variances,
#     - Independence of observations.

# 21. **What are the different types of ANOVA tests?**
#     - **One-way ANOVA**: Tests one independent variable.
#     - **Two-way ANOVA**: Tests two independent variables and their interaction.

# 22. **What is the F-test, and how does it relate to hypothesis testing?**
#     The F-test compares variances of two groups or tests multiple means in ANOVA. High F-values indicate group differences.

```

```python
import math
from scipy.stats import norm

def z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    """
    Perform a one-sample Z-test and interpret the results.

    Parameters:
        sample_mean (float): Mean of the sample.
        population_mean (float): Mean of the population.
        std_dev (float): Population standard deviation.
        sample_size (int): Size of the sample.
        alpha (float): Significance level (default is 0.05).

    Returns:
        None
    """
    # Calculate the Z-score
    standard_error = std_dev / math.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Interpretation
    print("Z-Test Results:")
    print(f"Sample Mean: {sample_mean}")
    print(f"Population Mean: {population_mean}")
    print(f"Standard Deviation: {std_dev}")
    print(f"Sample Size: {sample_size}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), "
              "we reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), "
              "we fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")

# Example usage
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 30
alpha = 0.05

z_test(sample_mean, population_mean, std_dev, sample_size, alpha)
```

```python
# Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
import numpy as np
from scipy.stats import ttest_1samp, norm

def simulate_and_test(population_mean=100, population_std=15, sample_size=30, alpha=0.05):
    """
    Simulate random data, perform hypothesis testing, and calculate the p-value.

    Parameters:
        population_mean (float): Mean of the population.
        population_std (float): Standard deviation of the population.
        sample_size (int): Size of the sample.
        alpha (float): Significance level (default is 0.05).

    Returns:
        None
    """
    # Simulate random data for the sample
    np.random.seed(42)  # For reproducibility
    sample = np.random.normal(population_mean, population_std, sample_size)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation

    print("Simulated Sample:")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Sample Size: {sample_size}")

    # Perform a one-sample t-test (used when population standard deviation is unknown)
    t_stat, p_value = ttest_1samp(sample, population_mean)

    # Interpret the results
    print("\nHypothesis Testing Results:")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), "
              "we reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), "
              "we fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")

# Example usage
simulate_and_test(population_mean=100, population_std=15, sample_size=30, alpha=0.05)
```

```python
import math
from scipy.stats import norm

def one_sample_z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    standard_error = std_dev / math.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# Example usage
one_sample_z_test(105, 100, 15, 30)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_two_tailed_z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    standard_error = std_dev / math.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error
    z_critical = norm.ppf(1 - alpha / 2)

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label="Normal Distribution")
    plt.axvline(-z_critical, color="red", linestyle="--", label="Critical Region")
    plt.axvline(z_critical, color="red", linestyle="--")
    plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color="red", alpha=0.3)
    plt.title("Two-tailed Z-Test Decision Region")
    plt.legend()
    plt.show()

    print(f"Z-Score: {z_score:.4f}")
    print(f"Critical Value: ±{z_critical:.4f}")
    if abs(z_score) > z_critical:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Example usage
visualize_two_tailed_z_test(105, 100, 15, 30)
```

```python
def type1_type2_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label="Null Distribution")
    plt.fill_between(x, y, where=(x > norm.ppf(1 - alpha)), color="red", alpha=0.3, label="Type 1 Error")
    plt.fill_between(x, y, where=(x < norm.ppf(beta)), color="blue", alpha=0.3, label="Type 2 Error")
    plt.axvline(norm.ppf(1 - alpha), color="red", linestyle="--", label="Critical Value")
    plt.axvline(norm.ppf(beta), color="blue", linestyle="--")
    plt.legend()
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

# Example usage
type1_type2_errors()
```

```python
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    t_stat, p_value = ttest_ind(group1, group2)

    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The means of the two groups are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The means of the two groups are not significantly different.")

# Example usage
independent_t_test([1, 2, 3, 4, 5], [5, 6, 7, 8, 9])
```

```python
from scipy.stats import ttest_rel

def paired_t_test(before, after, alpha=0.05):
    t_stat, p_value = ttest_rel(before, after)

    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The means of the paired samples are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The means of the paired samples are not significantly different.")

# Example usage
paired_t_test([1, 2, 3], [2, 2, 3])
```

```python
def compare_z_t_test():
    np.random.seed(42)
    data = np.random.normal(100, 15, 30)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    # Z-Test
    z_test_result = (sample_mean - 100) / (15 / math.sqrt(30))
    p_value_z = 2 * (1 - norm.cdf(abs(z_test_result)))

    # T-Test
    t_test_result, p_value_t = ttest_1samp(data, 100)

    print(f"Z-Test: Z-Score = {z_test_result:.4f}, P-Value = {p_value_z:.4f}")
    print(f"T-Test: T-Statistic = {t_test_result:.4f}, P-Value = {p_value_t:.4f}")

# Example usage
compare_z_t_test()
```

```python
def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    margin_of_error = norm.ppf((1 + confidence) / 2) * (sample_std / math.sqrt(n))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Example usage
confidence_interval([1, 2, 3, 4, 5])
```

```python
def calculate_margin_of_error(data, confidence=0.95):
    sample_std = np.std(data, ddof=1)
    n = len(data)
    critical_value = norm.ppf((1 + confidence) / 2)
    margin_of_error = critical_value * (sample_std / math.sqrt(n))

    print(f"Margin of Error: {margin_of_error:.4f}")

# Example usage
calculate_margin_of_error([12, 14, 15, 13, 14, 15])
```

```python
def bayesian_inference(prior, likelihood, marginal_likelihood):
    posterior = (likelihood * prior) / marginal_likelihood
    print(f"Posterior Probability: {posterior:.4f}")
    return posterior

# Example usage
prior = 0.01
likelihood = 0.8
marginal_likelihood = 0.02
bayesian_inference(prior, likelihood, marginal_likelihood)
```

```python
from scipy.stats import chi2_contingency

def chi_square_test(data):
    chi2, p_value, dof, expected = chi2_contingency(data)
    print(f"Chi-Square Statistic: {chi2:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print("Expected Frequencies:")
    print(expected)

    if p_value < 0.05:
        print("Reject the null hypothesis: The variables are not independent.")
    else:
        print("Fail to reject the null hypothesis: The variables are independent.")

# Example usage
data = [[50, 30], [20, 100]]
chi_square_test(data)
```

```python
def calculate_expected_frequencies(observed):
    total = np.sum(observed)
    row_totals = np.sum(observed, axis=1)
    col_totals = np.sum(observed, axis=0)
    expected = np.outer(row_totals, col_totals) / total
    print("Expected Frequencies:")
    print(expected)
    return expected

# Example usage
observed = np.array([[50, 30], [20, 100]])
calculate_expected_frequencies(observed)
```

```python
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):
    chi2_stat, p_value = chisquare(observed, f_exp=expected)
    print(f"Chi-Square Statistic: {chi2_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")

# Example usage
observed = [50, 30, 20]
expected = [40, 40, 20]
goodness_of_fit_test(observed, expected)
```

```python
from scipy.stats import f

def f_test_variance(data1, data2, alpha=0.05):
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1
    dfn = len(data1) - 1
    dfd = len(data2) - 1
    critical_value = f.ppf(1 - alpha, dfn, dfd)

    print(f"F-Statistic: {f_stat:.4f}")
    print(f"Critical Value: {critical_value:.4f}")
    if f_stat > critical_value:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Example usage
data1 = [1, 2, 3, 4, 5]
data2 = [2, 4, 6, 8, 10]
f_test_variance(data1, data2)
```

```python
from scipy.stats import f_oneway

def anova_test(*groups):
    f_stat, p_value = f_oneway(*groups)
    print(f"F-Statistic: {f_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis: At least one group mean is significantly different.")
    else:
        print("Fail to reject the null hypothesis: All group means are statistically similar.")

# Example usage
anova_test([1, 2, 3], [2, 3, 4], [4, 5, 6])
```

```python
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt

def f_test_and_visualize(data1, data2, alpha=0.05):
    # Calculate sample variances
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)

    # Ensure the larger variance is in the numerator
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(data1) - 1, len(data2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(data2) - 1, len(data1) - 1

    # Calculate the p-value
    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    print(f"F-Statistic: {f_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: Variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: Variances are not significantly different.")

    # Visualization of the F-distribution
    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, dfn, dfd)
    plt.plot(x, y, label=f"F-distribution (dfn={dfn}, dfd={dfd})")
    plt.axvline(f_stat, color='red', linestyle='--', label=f"F-statistic = {f_stat:.2f}")
    plt.title("F-Test Visualization")
    plt.xlabel("F-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

# Example usage
data1 = np.random.normal(10, 2, 50)
data2 = np.random.normal(10, 3, 50)
f_test_and_visualize(data1, data2)
```

```python
from scipy.stats import shapiro, levene

def check_anova_assumptions(data):
    # Assumption 1: Normality (Shapiro-Wilk Test)
    print("Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(data, 1):
        stat, p_value = shapiro(group)
        print(f"Group {i}: Stat={stat:.4f}, P-Value={p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i} does not follow a normal distribution.")
        else:
            print(f"Group {i} follows a normal distribution.")

    # Assumption 2: Equal Variance (Levene's Test)
    print("\nEquality of Variances (Levene's Test):")
    stat, p_value = levene(*data)
    print(f"Levene's Test Stat={stat:.4f}, P-Value={p_value:.4f}")
    if p_value < 0.05:
        print("The variances are significantly different (Violated assumption).")
    else:
        print("The variances are not significantly different (Assumption holds).")

# Example usage
data1 = np.random.normal(10, 2, 50)
data2 = np.random.normal(15, 3, 50)
data3 = np.random.normal(20, 1, 50)
check_anova_assumptions([data1, data2, data3])
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

def two_way_anova(data, factor1, factor2):
    # Prepare the data
    df = pd.DataFrame(data)

    # Fit the model
    model = ols('value ~ C(factor1) + C(factor2) + C(factor1):C(factor2)', data=df).fit()

    # Perform the ANOVA
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(anova_table)

    # Plot interaction between the two factors
    pd.plotting.scatter_matrix(df, alpha=0.8, figsize=(10, 8), diagonal='hist')
    plt.title("Interaction Plot")
    plt.show()

# Example usage (Create some sample data)
factor1 = np.random.choice(['A', 'B'], 100)
factor2 = np.random.choice(['X', 'Y'], 100)
values = np.random.normal(10, 5, 100)

data = {'factor1': factor1, 'factor2': factor2, 'value': values}
two_way_anova(data, 'factor1', 'factor2')
```
